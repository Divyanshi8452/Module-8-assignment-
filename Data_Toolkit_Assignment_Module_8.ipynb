2.  How does broadcasting work in NumPy?
--> Broadcasting in NumPy is a powerful feature that allows arithmetic operations on arrays of different shapes without explicitly copying data or reshaping arrays.
When operating on two arrays, NumPy compares their shapes element-wise from right to left:

-If the dimensions are equal, or
-If one of the dimensions is 1,
then they are compatible in that dimension.

If they are compatible in all dimensions, NumPy broadcasts the smaller array to match the larger one.

```python
#example for ques 2
#1. Scalar and Array
import numpy as np

a = np.array([1, 2, 3])
b = 10

print(a + b)

#2D Array and 1D Array
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([10, 20, 30])

print(A + B)

#3.Column Vector and Row Vector
A = np.array([[1],
              [2],
              [3]])

B = np.array([10, 20, 30])

B = B.reshape(1, 3)

print(A + B)
```

4. Explain the use of the groupby() method in Pandas.
--> The .groupby() method in Pandas is used to split your data into groups based on some criteria, then apply a function to each group, and finally combine the results.

```python
#example for ques 4
import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [50000, 55000, 60000, 52000, 70000]
}

df = pd.DataFrame(data)
```

17.  Explain the difference between apply() and map() in Pandas?
--> Apply():
1. Function application: apply() applies a function to each row or column of a DataFrame.
2. Flexible: apply() can be used with any function, including custom functions.
3. Axis specification: apply() allows specifying the axis (rows or columns) to apply the function.

Map():
1. Element-wise operation: map() applies a function element-wise to a Series.
2. Dictionary mapping: map() can also be used to map values from a dictionary.
3. Limited to Series: map() is limited to Series, whereas apply() can be used with DataFrames.

21. Why is NumPy's array slicing faster than Python's list slicing?
--> NumPy's array slicing is faster than Python's list slicing for several key reasons related to how data is stored and accessed in memory.
1. Contiguous Block of Memory
2. Fixed Data Types
3. Views, Not Copies
4. Low-Level Optimizations


```python
#example for ques 21
import numpy as np
import time

# NumPy array
a = np.arange(1_000_000)
start = time.time()
b = a[100_000:900_000]
print("NumPy slice time:", time.time() - start)

# Python list
c = list(range(1_000_000))
start = time.time()
d = c[100_000:900_000]
print("List slice time:", time.time() - start)
```

```python
''' 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.DataFrame({
    'Math': [90, 85, 80, 75, 70],
    'Science': [85, 80, 75, 70, 65],
    'English': [80, 75, 70, 65, 60]
})

# Compute correlation matrix
corr_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title('Correlation Heatmap')
plt.show()
```
